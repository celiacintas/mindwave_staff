```python
def mindwave_start():
    socket, socket_addr = connect_magic()
    if socket is None:
        print "No MindWave found"
        sys.exit(-1)
    print "Connected at %s" % socket_addr
    return socket
```

```python
socket = mindwave_start()
```

```python
recorder = TimeSeriesRecorder()
parser = ThinkGearParser(recorders=[recorder])
```
